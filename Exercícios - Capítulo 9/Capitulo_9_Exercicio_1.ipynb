```python
# Cálculo numérico para engenharia elétrica com PYTHON
# Capítulo 9: Derivação numérica
# Obtido de: https://github.com/PECCE-IFSC/CALCULO_NUMERICO_PYTHON/blob/main/C9_Formula%20de%20deriva%C3%A7%C3%A3o


from numpy import divide, diff
import matplotlib.pyplot as plt


t = [0, 0.5, 1, 2, 3]
V = [100, 50, 35, 9.5, 0.05]
d = divide(diff(V), diff(t))
C = 0.01
I = C * d
tt = [0.25, 0.75, 1.5, 2.5]

print(f'd = {d}\n')
print(f'I = {I}\n')

# Visualização
fig, ax = plt.subplots()
Corr = ax.plot(tt, I, linewidth=1, marker='d', label='Corrente')
ax2 = ax.twinx()  # cria nova escala no lado oposto do gráfico eixo y
Tens = ax2.plot(t, V, linewidth=1, marker='s', label='Tensão', linestyle='--', color='k')
lns = Tens + Corr
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Corrente (A)')
ax2.set_ylabel('Tensão (V)')
ax2.yaxis.grid(False)
plt.tight_layout()
plt.show()
```
