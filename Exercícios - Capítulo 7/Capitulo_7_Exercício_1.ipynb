1. Uma medição de aterramento chegou aos valores apresentados na tabela. Entendendo que esses valores são pouco precisos, trace a curva que melhor represente o comportamento das medições. Discuta sobre a correção dos dados as curvas traçadas.


|R (Ω)   | 1 | 1.1| 1.25 | 1.5 | 2 | 2.2 | 3.5 | 5 | 6 | 6,5 | 7  | 8 | 8.75 | 9.5 | 10  |
|-----------|---|----|------|-----|---|-----|-----|---|---|-----|----|---|------|-----|-----|
| I (A)  |10 | 9  | 8.9  | 7   | 6 | 5.5 | 4   | 2 |1.5| 1.1 |1.05| 1 | 1    | 1.1 | 0.95|

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 1.1, 1.25, 1.5, 2, 2.2, 3.5, 5, 6, 6.5, 7, 8, 8.75, 9.5, 10 ])
y = np.array([10, 9, 8.9, 7, 6, 5.5, 4, 2, 1.5, 1.1, 1.05, 1, 1, 1.1, 0.95 ])

# Usando Polyfit
n = 1 # grau do polinomio

c = np.polyfit(x, y, deg = n) # método dos mínimos quadrados

px = np.linspace(x[0], x[len(x)-1], 100)

py = np.polyval(c, px)

plt.plot(x, y, 'o', markersize = 5, label='data')
plt.plot(px, py, 'k-', linewidth = 1, label=f'Degree {n}')


# Usando Polyfit
n = 2 # grau do polinomio

c = np.polyfit(x, y, deg = n) # método dos mínimos quadrados

px = np.linspace(x[0], x[len(x)-1], 100)

py = np.polyval(c, px)

plt.plot(px, py, 'b-', linewidth = 1, label=f'Degree {n}')

# Usando Polyfit
n = 3 # grau do polinomio

c = np.polyfit(x, y, deg = n) # método dos mínimos quadrados

px = np.linspace(x[0], x[len(x)-1], 100)

py = np.polyval(c, px)

plt.plot(px, py, 'g-', linewidth = 1, label=f'Degree {n}')

# Usando Polyfit
n = 4 # grau do polinomio

c = np.polyfit(x, y, deg = n) # método dos mínimos quadrados

px = np.linspace(x[0], x[len(x)-1], 100)

py = np.polyval(c, px)

plt.plot(px, py, 'm-', linewidth = 1, label=f'Degree {n}')

# Usando Polyfit
n = 5 # grau do polinomio

c = np.polyfit(x, y, deg = n) # método dos mínimos quadrados

px = np.linspace(x[0], x[len(x)-1], 100)

py = np.polyval(c, px)

plt.grid(True)
plt.plot(px, py, 'r-', linewidth = 1, label=f'Degree {n}')
plt.xlabel('R (Ohms)')
plt.ylabel('I (A)')
plt.legend(loc = "upper right")

plt.show()
```

Considerando o conjunto de quinze pontos, observamos que o resultado obtido a partir de um polinômio de quinto grau já nos dá uma curva com o comportamento muito similar. Como o conjunto de pontos é impreciso, aumentar o grau do polinômio acarretaria em uma distorção da curva, que apesar de diminuir o erro relativo aos pontos, se afastaria mais do resultado real.
