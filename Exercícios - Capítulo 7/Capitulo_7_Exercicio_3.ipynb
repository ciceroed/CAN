```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as sp

def splines(interp, x, y):
  tck = sp. splrep(x, y)
  return sp.splev(interp, tck)

def pchip(interp, x, y):
  pchip = sp.PchipInterpolator(x, y)
  return pchip(interp)

# Vetores de dados
x = np.arange(11)
y = [80, -60, 40, -30, 20, -10, 5, -2.5, 1.25, -0.625, 0.3125]
interp = 1.5

print(f"Interpolação por Spline: y = {splines(interp, x, y):.4f}")
print(f"Interpolação por Pchip: y = {pchip(interp, x, y):.4f}")

# Plot

x_axis = np.linspace(min(x), max(x), 100)

y_axis = splines(x_axis, x, y)
plt.plot(x_axis, y_axis, "k-", label = "Spline")

y_axis = pchip(x_axis, x, y)
plt.plot(x_axis, y_axis, "r-", label = "Pchip")

plt.grid(True)
plt.plot(x, y, 'o', label = "Dados")
plt.legend()
plt.show()
```
